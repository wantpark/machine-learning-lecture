##### 전체 픽셀의 중앙값

````python
from numpy import asarray
from PIL import Image
from matplotlib import pyplot

# 이미지 읽기
image = pyplot.imread('Old Lower Shotover Bridge.jpg')

# 데이터 변환
pixels = asarray(image)
pixels = pixels.astype('float32')

# 전체 픽셀의 평균
mean = pixels.mean()
print('Mean: %.3f' % mean)
print('Min: %.3f, Max: %.3f' % (pixels.min(), pixels.max()))

# 중앙값 계산
pixels = pixels - mean

# 픽셀의 평균, 최대 그리고 최소
print('Mean: %.3f' % pixels.mean())
print('Min: %.3f, Max: %.3f' % (pixels.min(), pixels.max()))

# RGB 분리
r = pixels.copy()
g = pixels.copy()
b = pixels.copy()
r[:, :, 1] = 0
r[:, :, 2] = 0
g[:, :, 0] = 0
g[:, :, 2] = 0
b[:, :, 0] = 0
b[:, :, 1] = 0

image = Image.fromarray(pixels.astype('uint8'))
r = Image.fromarray(r.astype('uint8'))
g = Image.fromarray(g.astype('uint8'))
b = Image.fromarray(b.astype('uint8'))

f, ax = pyplot.subplots(2, 2)
ax[0, 0].imshow(image)
ax[0, 1].imshow(r)
ax[1, 0].imshow(g)
ax[1, 1].imshow(b)

pyplot.show()
````

##### 채널별 픽셀의 중앙값

````python
from numpy import asarray
from PIL import Image
from matplotlib import pyplot

# 이미지 읽기
image = pyplot.imread('Old Lower Shotover Bridge.jpg')

# 데이터 변환
pixels = asarray(image)
pixels = pixels.astype('float32')

# 각 픽셀의 평균
means = pixels.mean(axis=(0, 1), dtype='float64')
print('Mean: %s' % means)
print('Min: %s, Max: %s' % (pixels.min(axis=(0, 1)), pixels.max(axis=(0, 1))))

# 중앙값 계산
pixels = pixels - means

# 각 픽셀의 평균, 최대 그리고 최소
print('Mean: %s' % pixels.mean(axis=(0, 1)))
print('Min: %s, Max: %s' % (pixels.min(axis=(0, 1)), pixels.max(axis=(0, 1))))

# RGB 분리
r = pixels.copy()
g = pixels.copy()
b = pixels.copy()
r[:, :, 1] = 0
r[:, :, 2] = 0
g[:, :, 0] = 0
g[:, :, 2] = 0
b[:, :, 0] = 0
b[:, :, 1] = 0

image = Image.fromarray(pixels.astype('uint8'))
r = Image.fromarray(r.astype('uint8'))
g = Image.fromarray(g.astype('uint8'))
b = Image.fromarray(b.astype('uint8'))

f, ax = pyplot.subplots(2, 2)
ax[0, 0].imshow(image)
ax[0, 1].imshow(r)
ax[1, 0].imshow(g)
ax[1, 1].imshow(b)

pyplot.show()
````
