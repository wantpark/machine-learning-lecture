##### 표준화

```
전체 픽셀
```

````python
from numpy import asarray
from PIL import Image

# 이미지 읽기
image = Image.open('Old Lower Shotover Bridge.jpg')

# 데이터 변환
pixels = asarray(image)
pixels = pixels.astype('float32')
# 전체 픽셀의 평균과 표준편차
print('Mean: %.3f, Standard Deviation: %.3f' % (pixels.mean(), pixels.std()))

# 표준화
pixels = (pixels - pixels.mean()) / pixels.std()
# 전체 픽셀의 평균과 표준편차
print('Mean: %.3f, Standard Deviation: %.3f' % (pixels.mean(), pixels.std()))
# 전체 픽셀의 최소와 최대
print('Min: %s, Max: %s' % (pixels.min(), pixels.max()))
````

###### 표준화

```
전체 픽셀
 - 범위 : -1 ~ 1 사이를 0 ~ 1로 변경
```

````python
from numpy import asarray
from numpy import clip
from PIL import Image

# 이미지 읽기
image = Image.open('Old Lower Shotover Bridge.jpg')

# 데이터 변환
pixels = asarray(image)
pixels = pixels.astype('float32')
# 전체 픽셀의 평균과 표준편차
print('Mean: %.3f, Standard Deviation: %.3f' % (pixels.mean(), pixels.std()))

# 표준화
pixels = (pixels - pixels.mean()) / pixels.std()

# -1 ~ 1 범위만
pixels = clip(pixels, -1.0, 1.0)
# 범위를 0 ~ 1로 변경
pixels = (pixels + 1.0) / 2.0
# 전체 픽셀의 평균과 표준편차
print('Mean: %.3f, Standard Deviation: %.3f' % (pixels.mean(), pixels.std()))
# 전체 픽셀의 최소와 최대
print('Min: %.3f, Max: %.3f' % (pixels.min(), pixels.max()))
````

##### 표준화

```
채널별로
```

````python
from numpy import asarray
from PIL import Image

# 이미지 읽기
image = Image.open('Old Lower Shotover Bridge.jpg')

# 데이터 변환
pixels = asarray(image)
pixels = pixels.astype('float32')
# 채널별 평균과 표준편차
mean = pixels.mean(axis=(0, 1), dtype='float64')
std = pixels.std(axis=(0, 1), dtype='float64')
print('Mean: %s, Standard Deviation: %s' % (mean, std))
# 채널별 최소와 최대
print('Min: %s, Max: %s' % (pixels.min(axis=(0, 1)), pixels.max(axis=(0, 1))))

# 표준화
pixels = (pixels - mean) / std
# 채널별 평균과 표준편차
mean = pixels.mean(axis=(0, 1), dtype='float64')
std = pixels.std(axis=(0, 1), dtype='float64')
print('Mean: %s, Standard Deviation: %s' % (mean, std))
# 채널별 최소와 최대
print('Min: %s, Max: %s' % (pixels.min(axis=(0, 1)), pixels.max(axis=(0, 1))))
````
