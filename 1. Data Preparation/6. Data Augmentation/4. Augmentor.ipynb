##### Initialising a pipeline

```python
import Augmentor

path_to_data = "images"
output = "images/output"

# Create a pipeline
p = Augmentor.Pipeline(path_to_data)
```

##### Adding operations to a pipeline

```python
# Defining augmentation parameters
p.flip_left_right(probability=0.4) 
p.flip_top_bottom(probability=0.8)
p.rotate(probability=0.5, max_left_rotation=5, max_right_rotation=10)
p.skew(0.4, 0.5) 
p.zoom(probability = 0.2, min_factor = 1.1, max_factor = 1.5) 
```

##### Executing a pipeline

```python
import os

for f in os.listdir(output):
    os.remove(os.path.join(output, f))

# Here we sample 36 images from the pipeline.

# It is often useful to use scientific notation for specify
# large numbers with trailing zeros.
num_of_samples = 36

# Now we can sample from the pipeline:
p.sample(num_of_samples)
```

##### View

```python
import imageio
from matplotlib import pyplot
from pathlib import Path

images = list()

for file in Path(output).iterdir():
    im = imageio.v2.imread(file)
    images.append(im)

pyplot.figure(figsize=(15, 11))

for i in range(1, 37):
    pyplot.subplot(6, 6, i)
    pyplot.axis('off')
    pyplot.tight_layout()
    pyplot.imshow(images[i - 1])
```
