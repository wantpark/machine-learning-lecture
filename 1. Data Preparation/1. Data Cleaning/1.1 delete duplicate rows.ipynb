#### IGN 비디오 게임 데이터

```
중복값이 있는 특성 제거
```

````python
from pandas import read_csv

# IGN 비디오 게임 데이터
path = 'https://raw.githubusercontent.com/john7obed/ign_games_of_20_years/master/ign.csv'
# 데이터 세트 로드
df = read_csv(path)

# 데이터 분할
df = df.iloc[:, 1:]
# 데이터 개수와 특성 크기
print(df.shape)

# 중복값 확인
dups = df.duplicated()
# 중복값이 하나라도 있으면 True, 그렇지 않으면 False 
print(dups.any(), len(df[dups]))

# 중복값이 있는 데이터
print(df[dups])

# 중복값 제거
df.drop_duplicates(inplace=True)
# 데이터 개수와 특성 크기
print(df.shape)
````

#### 게임 데이터

```
중복값이 있는 특성 제거
```

````python
from pandas import read_csv

# 스팀 게임 데이터
path = 'https://raw.githubusercontent.com/lix229/steamer/master/data/steam-200k.csv'
# 데이터 세트 로드
df = read_csv(path, header=None)
# 데이터 개수와 특성 크기
print(df.shape)

# 중복값 확인
dups = df.duplicated()
# 중복값이 하나라도 있으면 True, 그렇지 않으면 False 
print(dups.any(), len(df[dups]))

# 중복값이 있는 데이터
print(df[dups])

# 중복값 제거
df.drop_duplicates(inplace=True)
# 데이터 개수와 특성 크기
print(df.shape)
````
